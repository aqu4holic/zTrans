```python
import pandas as pd
import numpy as np

seed = 18022004
np.random.seed(seed)
```

```python
data_prefix = "data"

repo_df = pd.read_csv(f'{data_prefix}/classified_res.csv')
```

```python
a = repo_df[repo_df['startCommit'] == repo_df['endCommit']]
```

```python
data_length = 500
sampled_data = a.sample(n = data_length, random_state = seed)
```

```python
sampled_data
```

```python
sampled_data.to_parquet(f'{data_prefix}/500_sampled_raw.parquet')
```
