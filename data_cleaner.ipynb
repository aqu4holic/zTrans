```python
import pandas as pd
import numpy as np

seed = 18022004
np.random.seed(seed)
```

```python
data_prefix = 'data'

# filtered startCommit == endCommit

data_name = 'classified_res.csv'

repo_df = pd.read_csv(f'{data_prefix}/{data_name}')
```

```python
# add repoOwner and repoID column

repo_df['repoOwner'] = None
repo_df['repoSplitName'] = None

for id in range(len(repo_df)):
    sample = repo_df.iloc[id]

    repo_owner, repo_id = sample['repoName'].split('_', 1)

    repo_df.at[id, 'id'] = id
    repo_df.at[id, 'repoOwner'] = repo_owner
    repo_df.at[id, 'repoSplitName'] = repo_id

repo_df.to_parquet(f'{data_prefix}/original.parquet')
```

```python
data_name = 'original.parquet'

repo_df = pd.read_parquet(f'{data_prefix}/{data_name}', engine = 'pyarrow')
```

```python
import os

error_repos = set()

for repo_name in repo_df['repoName']:
    if (not os.path.exists(f'{data_prefix}/repos/{repo_name}')):
        error_repos.add(repo_name)
```
