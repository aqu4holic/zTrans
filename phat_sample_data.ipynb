# init

```python
import pandas as pd
import numpy as np

seed = 18022004
np.random.seed(seed)
```

```python
data_prefix: str = 'data'
repo_prefix: str = f'{data_prefix}/repos'

data_name = 'official_original.parquet'

original_df = pd.read_parquet(f'{data_prefix}/{data_name}', engine = 'pyarrow')
```

# add 2 cols

```python
original_df['repoOwner'] = None
original_df['repoSplitName'] = None

for id in range(len(original_df)):
    sample = original_df.iloc[id]

    repo_name = sample['repoName']

    repo_owner, repo_split_name = repo_name.split('_', 1)

    original_df.at[id, 'repoOwner'] = repo_owner
    original_df.at[id, 'repoSplitName'] = repo_split_name
```

```python
original_df.to_parquet(f'{data_prefix}/{data_name}', engine = 'pyarrow')
```

# sampling

## sampling all

```python
sample_cnt = 2000

sampled_all_df = original_df.sample(n = sample_cnt, random_state = seed)
```

```python
sampled_all_df.to_parquet(f'{data_prefix}/{sample_cnt}_sample_no_code.parquet')
```

```python
def df_reset_index(df):
    df.reset_index(drop = True, inplace = True)
    df.loc[:, 'id'] = df.index
    return df
```

```python
log_sample_df = log_df.sample(n=500, random_state=seed)
test_sample_df = test_unit_df.sample(n=500, random_state=seed)
others_sample_df = remain_df.sample(n=500, random_state=seed)
```
