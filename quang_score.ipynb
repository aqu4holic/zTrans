````python
import re

def filter_code_prediction(text: str) -> str:
    extracted_text = re.search(r"```(?:\w+)?\n(.*?)```", text, re.DOTALL)
    if extracted_text:
        result = extracted_text.group(1).lstrip("\n").replace("```", "")
        return result
````

````python
result['predicted'] = result['predicted'].apply(filter_code_prediction)
````
