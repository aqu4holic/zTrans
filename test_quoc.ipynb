```python
import difflib
import re

def find_changed_method_signatures(start_code, end_code):
    start_lines = start_code.splitlines()
    end_lines = end_code.splitlines()
    
    # Sử dụng difflib để so sánh và tìm sự thay đổi
    diff_obj = difflib.ndiff(start_lines, end_lines)
    changes = list(diff_obj)
    
    # Biểu thức regex để tìm tên phương thức (public/private/protected + tên phương thức)
    method_regex = re.compile(r'\b(public|private|protected)?\s+\w+\s+(\w+)\s*\(')
    
    # Duyệt qua sự thay đổi và xác định các phương thức liên quan
    changed_methods = set()
    current_method = ""
    for line in range(len(changes)):
        # Kiểm tra nếu dòng là định nghĩa phương thức mới
        method_match = method_regex.search(changes[line])
        if method_match:
            current_method = ""
            current_method += changes[line] # Lấy tên phương thức
            while '}' not in changes[line]:
                if changes[line].strip() == '}':
                    current_method += changes[line]
                    break  # Thoát vòng lặp khi gặp dòng chỉ có '}'
                current_method += changes[line]
                current_method += '\n'
                line+=1
        # Kiểm tra nếu dòng thay đổi
        if changes[line].startswith('+ ') or changes[line].startswith('- '):
            if current_method:
                changed_methods.add(current_method)
    
    return list(changed_methods)



# data['method_signature_changes'] = data.apply(lambda row: find_changed_method_signatures(row['startCode'], row['endCode']), axis=1)
```
